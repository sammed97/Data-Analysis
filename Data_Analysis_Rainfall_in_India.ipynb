```python
#Highest rainfall over given years
df_max_annual_rainfall=df[df['ANNUAL']==df['ANNUAL'].max()]
df_max_annual_rainfall
```

In 1948 Arunachal Pradesh recieved highest annual rainfall ever

```python
#Minimum rainfall over given years
df_min_annual_rainfall=df[df['ANNUAL']==df['ANNUAL'].min()]
df_min_annual_rainfall
```

In 1918 West Rajasthan recieved lowest rainfall

```python
max_rainfall_divisionwise=df.groupby('SUBDIVISION')['ANNUAL'].max()
max_rainfall_divisionwise.reset_index().sort_values(by='ANNUAL',ascending=False)
```

```python
max_rainfall_yearwise=df.groupby('YEAR')['ANNUAL'].max()
max_rainfall_yearwise.reset_index()
```

```python
max_rainfall_yearwise_df=pd.DataFrame(max_rainfall_yearwise.reset_index())
max_rainfall_yearwise_df.rename(columns={'ANNUAL':'Max Annual Rainfall'},inplace=True)

max_rainfall_yearwise_df
```

```python
quartewise_rainfall_1=pd.DataFrame(round(df.groupby('SUBDIVISION')['Jan-Feb','Latitude','Longitude'].mean(),2))
quartewise_rainfall_2=pd.DataFrame(round(df.groupby('SUBDIVISION')['Mar-May','Latitude','Longitude'].mean(),2))
quartewise_rainfall_3=pd.DataFrame(round(df.groupby('SUBDIVISION')['June-September','Latitude','Longitude'].mean(),2))
quartewise_rainfall_4=pd.DataFrame(round(df.groupby('SUBDIVISION')['Oct-Dec','Latitude','Longitude'].mean(),2))
```

```python
quartewise_rainfall_1=pd.DataFrame(round(df.groupby('SUBDIVISION')['Jan-Feb','Latitude','Longitude'].mean(),2))
```

```python
quartewise_rainfall_1.sort_values(by='Jan-Feb',ascending=False)
```

```python
fig, size = plt.subplots(figsize=(6, 8))
sns.scatterplot(data=quartewise_rainfall_1, x="Longitude", y="Latitude", hue="Jan-Feb", size="Jan-Feb")
plt.title('Jan-Feb Average Rainfall')
```

```python
quartewise_rainfall_2.sort_values(by='Mar-May',ascending=False)
```

```python
fig, size = plt.subplots(figsize=(6, 8))
sns.scatterplot(data=quartewise_rainfall_2, x="Longitude", y="Latitude", hue="Mar-May", size="Mar-May")
plt.title('Mar-May Average Rainfall')
```

```python
quartewise_rainfall_3.sort_values(by='June-September',ascending=False)
```

```python
fig, size = plt.subplots(figsize=(6, 8))
sns.scatterplot(data=quartewise_rainfall_3, x="Longitude", y="Latitude", hue="June-September", size="June-September")
plt.title("'June-September' Average Rainfall")
```

```python
quartewise_rainfall_4.sort_values(by='Oct-Dec',ascending=False)
```

```python
fig, size = plt.subplots(figsize=(6, 8))
sns.scatterplot(data=quartewise_rainfall_4, x="Longitude", y="Latitude", hue="Oct-Dec", size="Oct-Dec")
plt.title("'Oct-Dec' Average Rainfall")
```

```python
#Highest rainfall over given years
df_max_annual_rainfall=df[df['ANNUAL']==df['ANNUAL'].max()]
df_max_annual_rainfall
```

In 1948 Arunachal Pradesh recieved highest annual rainfall ever

```python
#Minimum rainfall over given years
df_min_annual_rainfall=df[df['ANNUAL']==df['ANNUAL'].min()]
df_min_annual_rainfall
```

In 1918 West Rajasthan recieved lowest rainfall

```python
max_rainfall_divisionwise=df.groupby('SUBDIVISION')['ANNUAL'].max()
max_rainfall_divisionwise.reset_index().sort_values(by='ANNUAL',ascending=False)
```

```python
max_rainfall_yearwise=df.groupby('YEAR')['ANNUAL'].max()
max_rainfall_yearwise.reset_index()
```

```python
max_rainfall_yearwise_df=pd.DataFrame(max_rainfall_yearwise.reset_index())
max_rainfall_yearwise_df.rename(columns={'ANNUAL':'Max Annual Rainfall'},inplace=True)

max_rainfall_yearwise_df
```

```python
quartewise_rainfall=pd.DataFrame(round(df.groupby('SUBDIVISION')['Jan-Feb','Mar-May','June-September','Oct-Dec'].mean(),2))
```
